**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
# Coordenadas de la escuela y el hospital
school_coords = (-118, 37)
hospital_coords = (-122, 34)

# FunciÃ³n para calcular la distancia euclidiana
def euclidean_distance(lat1, lon1, lat2, lon2):
    return np.sqrt((lat2 - lat1)**2 + (lon2 - lon1)**2)

# Calcular la distancia de cada casa a la escuela y al hospital
df_housing["distance_to_school"] = df_housing.apply(lambda row: euclidean_distance(
    row["latitude"], row["longitude"], school_coords[1], school_coords[0]), axis=1)

df_housing["distance_to_hospital"] = df_housing.apply(lambda row: euclidean_distance(
    row["latitude"], row["longitude"], hospital_coords[1], hospital_coords[0]), axis=1)
```

```python
# Clasificar si las casas estÃ¡n cerca de una escuela o un hospital
df_housing["close_to_school_or_hospital"] = (df_housing["distance_to_school"] < 0.50) | (df_housing["distance_to_hospital"] < 0.50)
```

```python
import scipy.stats as stats

# Dividir las casas en cercanas y lejanas
close_houses = df_housing[df_housing["close_to_school_or_hospital"] == True]["median_house_value"]
far_houses = df_housing[df_housing["close_to_school_or_hospital"] == False]["median_house_value"]

# Prueba ANOVA
f_stat, p_value = stats.f_oneway(close_houses, far_houses)
f_stat, p_value
```

```python
H0 = "H0 (There is no significant difference in prices)."
H1 = "H1 (There is a significant difference in prices)."
alpha = 0.05
if p_value < alpha:
    print(f"Result: {H1} (p-value = {p_value:.4f}).")
else:
    print(f"Result: {H0} (p-value = {p_value:.4f}).")
```
